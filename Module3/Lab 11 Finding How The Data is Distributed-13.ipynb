### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
print(df.columns)
print('')
print(df.dtypes)
print('')
print(df.info())
print('')
print(df.shape)
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)

import pandas as pd

# Numeric and categorical column lists
num_cols = df.select_dtypes(include=['int64','float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values left:", df.isnull().sum().sum())
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
# Employment distribution
print("Employment Distribution:")
print(df['Employment'].value_counts(dropna=False))
print("\n")

# Job Satisfaction distribution
print("JobSat Distribution:")
print(df['JobSat'].value_counts(dropna=False))
print("\n")

# Years of Professional Coding Experience distribution
print("YearsCodePro Distribution:")
print(df['YearsCodePro'].value_counts(dropna=False).sort_index())
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Count values of JobSat
job_sat_counts = df['JobSat'].value_counts(dropna=False)

# Use Seaborn "Set2" palette
colors = sns.color_palette("Set2", len(job_sat_counts))

# Pie chart
plt.figure(figsize=(7,7))
plt.pie(
    job_sat_counts,
    labels=job_sat_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors
)
plt.title("Job Satisfaction Distribution", fontsize=14)
plt.show()

```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Split and explode LanguageHaveWorkedWith
have_df = df.dropna(subset=['LanguageHaveWorkedWith']).copy()
have_df = have_df.assign(Language=have_df['LanguageHaveWorkedWith'].str.split(';'))
have_df = have_df.explode('Language')
have_df['Language'] = have_df['Language'].str.strip()

# Count frequencies
have_counts = have_df['Language'].value_counts().reset_index()
have_counts.columns = ['Language', 'HaveWorkedWith']

# Split and explode LanguageWantToWorkWith
want_df = df.dropna(subset=['LanguageWantToWorkWith']).copy()
want_df = want_df.assign(Language=want_df['LanguageWantToWorkWith'].str.split(';'))
want_df = want_df.explode('Language')
want_df['Language'] = want_df['Language'].str.strip()

# Count frequencies
want_counts = want_df['Language'].value_counts().reset_index()
want_counts.columns = ['Language', 'WantToWorkWith']

# Merge both counts
lang_compare = pd.merge(have_counts, want_counts, on='Language', how='outer').fillna(0)

# Take top 10 most common (based on sum)
lang_compare['Total'] = lang_compare['HaveWorkedWith'] + lang_compare['WantToWorkWith']
top_langs = lang_compare.sort_values(by='Total', ascending=False).head(10)

# Plot grouped bar chart
plt.figure(figsize=(12,6))
top_langs_melted = top_langs.melt(id_vars='Language', value_vars=['HaveWorkedWith', 'WantToWorkWith'],
                                  var_name='Category', value_name='Count')

sns.barplot(data=top_langs_melted, x='Language', y='Count', hue='Category', palette='viridis')
plt.title("Comparison of Programming Languages: Worked With vs Want to Work With")
plt.xticks(rotation=45)
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cross-tab between Country and RemoteWork
remote_ct = pd.crosstab(df['Country'], df['RemoteWork'])

# Pick top 10 countries with most respondents
top_countries = df['Country'].value_counts().head(10).index
remote_ct = remote_ct.loc[top_countries]

# Convert to long form for seaborn
remote_melted = remote_ct.reset_index().melt(id_vars='Country', var_name='RemoteWork', value_name='Count')

# Grouped bar chart
plt.figure(figsize=(12,6))
sns.barplot(data=remote_melted, x='Country', y='Count', hue='RemoteWork', palette='viridis')
plt.title("Distribution of Remote Work Preferences by Country (Top 10)")
plt.xticks(rotation=45)
plt.ylabel("Number of Respondents")
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Fix YearsCodePro
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0,
    'More than 50 years': 51
})
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop missing values
clean_df = df[['YearsCodePro', 'JobSat']].dropna()

# Correlation
pearson_corr, _ = pearsonr(clean_df['YearsCodePro'], clean_df['JobSat'])
spearman_corr, _ = spearmanr(clean_df['YearsCodePro'], clean_df['JobSat'])

print(f"Pearson correlation: {pearson_corr:.3f}")
print(f"Spearman correlation: {spearman_corr:.3f}")


```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Cross-tabulation
crosstab = pd.crosstab(df['EdLevel'], df['Employment'], normalize='index')  # normalize per education level
print(crosstab.head())

# Stacked bar plot
crosstab.plot(kind='bar', stacked=True, figsize=(12,6), colormap='viridis')

plt.title("Employment Status by Education Level")
plt.xlabel("Education Level")
plt.ylabel("Proportion of Respondents")
plt.legend(title="Employment", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
