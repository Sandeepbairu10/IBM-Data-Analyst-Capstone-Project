**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Demo: Running an SQL Query


Count the number of rows in the table named 'main'


```python
QUERY = """
SELECT COUNT(*) 
FROM main
"""
df = pd.read_sql_query(QUERY, conn)
df.head()
```

## Demo: Listing All Tables


To view the names of all tables in the database:


```python
QUERY = """
SELECT name as Table_Name FROM sqlite_master 
WHERE type = 'table'
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Running a Group By Query
    
For example, you can group data by a specific column, like Age, to get the count of respondents in each age group:


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
pd.read_sql_query(QUERY, conn)
```

## Demo: Describing a table

Use this query to get the schema of a specific table, main in this case:


```python
table_name = 'main'

QUERY = """
SELECT sql FROM sqlite_master 
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY, conn)
print(df.iat[0,0])
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here
import matplotlib.pyplot as plt

# Reconnect and read data
conn = sqlite3.connect("survey-data.sqlite")
df = pd.read_sql_query("SELECT CompTotal FROM main", conn)
conn.close()

# Convert to numeric
df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')

# Drop NaNs
comp_total = df['CompTotal'].dropna()

# Plot histogram
plt.figure(figsize=(8,6))
plt.hist(comp_total, bins=50, edgecolor="black", color="lightgreen")
plt.title("Histogram of Total Compensation (CompTotal)")
plt.xlabel("Total Compensation (USD)")
plt.ylabel("Frequency")
plt.show()

```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import pandas as pd

# Load Age column from SQLite
conn = sqlite3.connect("survey-data.sqlite")
df = pd.read_sql_query("SELECT Age FROM main", conn)
conn.close()

# Mapping Age categories to numeric midpoints
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

# Apply mapping
df['Age_numeric'] = df['Age'].map(age_mapping)

# Drop NaNs
age_data = df['Age_numeric'].dropna()

# Plot boxplot
plt.figure(figsize=(6,5))
sns.boxplot(x=age_data, color="lightblue")
plt.title("Boxplot of Age")
plt.xlabel("Age")
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
import pandas as pd

# Load Age and YearsCodePro from SQLite
conn = sqlite3.connect("survey-data.sqlite")
df = pd.read_sql_query("SELECT Age, YearsCodePro FROM main", conn)
conn.close()

# Mapping Age categories to numeric midpoints
age_mapping = {
    'Under 18 years old': 17,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

# Apply mapping
df['Age_numeric'] = df['Age'].map(age_mapping)

# Convert YearsCodePro to numeric
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0,
    'More than 50 years': 51
})
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Drop NaNs
scatter_df = df[['Age_numeric', 'YearsCodePro']].dropna()

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=scatter_df, x="Age_numeric", y="YearsCodePro", alpha=0.6)
plt.title("Scatter Plot of Age vs Work Experience (YearsCodePro)")
plt.xlabel("Age")
plt.ylabel("Work Experience (Years)")
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here
import matplotlib.pyplot as plt

# Map Age groups to numeric midpoints
age_map = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 29,
    "35-44 years old": 39,
    "45-54 years old": 49,
    "55-64 years old": 59,
    "65 years or older": 70
}

df['Age_numeric'] = df['Age'].map(age_map)

# Convert categorical columns to numeric codes for plotting
df['TimeNum'] = df['TimeSearching'].astype('category').cat.codes
df['FrustrationNum'] = df['Frustration'].astype('category').cat.codes

# Create bubble plot
plt.figure(figsize=(12, 7))
scatter = plt.scatter(
    df['TimeNum'], 
    df['FrustrationNum'], 
    s=df['Age_numeric'] * 5,   # bubble size
    c=df['Age_numeric'], 
    cmap='viridis', 
    alpha=0.6, 
    edgecolors='w', 
    linewidth=0.5
)

# Set category labels on axes
plt.xticks(df['TimeNum'].unique(), df['TimeSearching'].astype('category').cat.categories, rotation=45, ha='right')
plt.yticks(df['FrustrationNum'].unique(), df['Frustration'].astype('category').cat.categories)

plt.xlabel("Time Searching")
plt.ylabel("Frustration")
plt.title("Bubble Plot of Time Searching vs Frustration (Bubble Size = Age)")
plt.colorbar(scatter, label="Age midpoint")

plt.tight_layout()
plt.show()




```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Drop missing values
db_series = df['DatabaseWantToWorkWith'].dropna()

# Split multiple responses (semicolon-separated) into separate entries
db_exploded = db_series.str.split(';').explode()

# Count frequencies
db_counts = db_exploded.value_counts()

# Get top 5
top5 = db_counts.head(5)

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top5, 
    labels=top5.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=plt.cm.Paired.colors
)

plt.title("Top 5 Databases Respondents Want to Work With Next Year")
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# --- Step 1: Map Age ranges to numeric (midpoint) ---
age_mapping = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 30,
    "35-44 years old": 40,
    "45-54 years old": 50,
    "55-64 years old": 60,
    "65 years or older": 70
}
df['Age_numeric'] = df['Age'].map(age_mapping)

# --- Step 2: Map Time ranges to numeric (minutes) ---
time_mapping = {
    "Less than 15 minutes a day": 7.5,
    "15-30 minutes a day": 22.5,
    "30-60 minutes a day": 45,
    "1-2 hours a day": 90,
    "2-3 hours a day": 150,
    "3-4 hours a day": 210,
    "4-5 hours a day": 270,
    "More than 5 hours a day": 330
}

df['TimeSearching_num'] = df['TimeSearching'].map(time_mapping)
df['TimeAnswering_num'] = df['TimeAnswering'].map(time_mapping)

# --- Step 3: Filter age group 30–35 ---
age_group = df[(df['Age_numeric'] >= 30) & (df['Age_numeric'] <= 35)]

# --- Step 4: Calculate medians ---
median_searching = age_group['TimeSearching_num'].median()
median_answering = age_group['TimeAnswering_num'].median()

# --- Step 5: Create stacked bar chart ---
plt.figure(figsize=(6, 6))
plt.bar(
    ["30-35 Age Group"], 
    median_searching, 
    label="Median Time Searching"
)
plt.bar(
    ["30-35 Age Group"], 
    median_answering, 
    bottom=median_searching, 
    label="Median Time Answering"
)

plt.ylabel("Median Time (minutes)")
plt.title("Median Time Searching & Answering for Age Group 30-35")
plt.legend()
plt.show()


```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# --- Step 1: Map Age ranges to numeric (midpoint) ---
age_mapping = {
    "Under 18 years old": 17,
    "18-24 years old": 21,
    "25-34 years old": 30,
    "35-44 years old": 40,
    "45-54 years old": 50,
    "55-64 years old": 60,
    "65 years or older": 70
}
df['Age_numeric'] = df['Age'].map(age_mapping)

# --- Step 2: Filter respondents between 45 and 60 years ---
age_group = df[(df['Age_numeric'] >= 45) & (df['Age_numeric'] <= 60)]

# --- Step 3: Group by Age and calculate median CompTotal ---
median_comp = age_group.groupby('Age')['CompTotal'].median().reset_index()

# --- Step 4: Plot ---
plt.figure(figsize=(7,5))
plt.bar(median_comp['Age'], median_comp['CompTotal'], color="skyblue")
plt.ylabel("Median Compensation (CompTotal)")
plt.xlabel("Age Group")
plt.title("Median Compensation for Ages 45 to 60")
plt.xticks(rotation=45)
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
import matplotlib.pyplot as plt

# Count frequency of each MainBranch category
branch_counts = df['MainBranch'].value_counts()

# Plot horizontal bar chart
plt.figure(figsize=(10,6))
branch_counts.plot(kind='barh', color='teal', alpha=0.7)

plt.xlabel("Number of Respondents")
plt.ylabel("MainBranch")
plt.title("Distribution of MainBranch")
plt.gca().invert_yaxis()   # so the largest bar appears at the top
plt.show()
```
