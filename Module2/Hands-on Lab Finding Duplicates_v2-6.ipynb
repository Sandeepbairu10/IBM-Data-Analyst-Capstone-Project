### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here
dup_groups = df[df.duplicated(subset=["MainBranch", "Employment", "RemoteWork"], keep=False)]

dup_analysis = (
    dup_groups.groupby(["MainBranch", "Employment", "RemoteWork"]).agg(lambda x: x.nunique()))

print(dup_analysis)
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
## Write your code here


import pandas as pd
import matplotlib.pyplot as plt

# Example: load your dataset
# df = pd.read_csv("your_file.csv")

# Step 1: Filter duplicate rows based on MainBranch, Employment, RemoteWork
duplicates = df[df.duplicated(subset=["MainBranch", "Employment", "RemoteWork"], keep=False)]

# Step 2: Function to group small categories
def group_small_categories(series, top_n=10):
    top_categories = series.value_counts().nlargest(top_n)
    other_count = series.value_counts().iloc[top_n:].sum()
    if other_count > 0:
        top_categories["Other"] = other_count
    return top_categories

# Step 3: Group Country and Employment counts
country_counts = group_small_categories(duplicates["Country"], top_n=10)
employment_counts = group_small_categories(duplicates["Employment"], top_n=8)

# Step 4: Bar chart for Country distribution
plt.figure(figsize=(10, 6))
country_counts.plot(kind="bar", color="#4C9AFF", edgecolor="black")
plt.title("Duplicate Rows by Country (Top 10 + Other)")
plt.xlabel("Country")
plt.ylabel("Number of Duplicates")
plt.xticks(rotation=45, ha="right")
for i, val in enumerate(country_counts):
    plt.text(i, val + 0.5, str(val), ha='center', fontsize=9)
plt.tight_layout()
plt.show()

# Step 5: Pie chart for Employment distribution
plt.figure(figsize=(8, 8))
employment_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90, colors=plt.cm.Set3.colors)
plt.title("Duplicate Rows by Employment (Grouped)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
