### 2. Explore the Dataset

```python
## Write your code here
print(df.info())
print(df.describe())
```

### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
missing_values = df.isnull().sum()
missing_values
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here
# Create a figure for the heatmap
plt.figure(figsize=(12, 6))

# Use Seaborn heatmap to show missing values
sns.heatmap(df.isnull(), 
            cbar=False,       # Remove color bar
            cmap='viridis',   # Color map
            yticklabels=False # Hide row labels for cleaner view
           )

plt.title("Missing Values Heatmap", fontsize=16)
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
# Count missing values
missing_count = df["RemoteWork"].isnull().sum()


print(f"Missing values in 'RemoteWork': {missing_count} rows")
```

```python
df.head()
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here

# Find most frequent value (mode)
most_frequent = df["Employment"].mode()[0]

most_frequent
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Fill missing values with mode
df["Employment"].fillna(most_frequent, inplace=True)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
import matplotlib.pyplot as plt


def group_small_categories(series, top_n=10):
    top_categories = series.value_counts().nlargest(top_n)
    other_count = series.value_counts().iloc[top_n:].sum()
    if other_count > 0:
        top_categories["Other"] = other_count
    return top_categories
    
employment_counts = group_small_categories(df["Employment"], top_n=8)


plt.figure(figsize=(10, 6))
employment_counts.plot(kind="bar", color="#4C9AFF", edgecolor="black")

# Title & labels
plt.title("Employment Distribution After Imputation", fontsize=14)
plt.xlabel("Employment Type", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Display value labels above bars
for i, val in enumerate(employment_counts):
    plt.text(i, val + 0.5, str(val), ha='center', fontsize=9)

plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
